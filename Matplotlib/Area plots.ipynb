# Area Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Canada.csv')
```

```python
df.head()
```

```python
df.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
```

```python
year = list(map(str, range(1980,2014)))
df.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
df_top5 = df.head()
df_top5 = df_top5[year].transpose()
```

```python
df_top5.plot(kind='area')
plt.title('Immigration trends of top 5 countries')
plt.ylable('Number of immigrants')
plt.xlable('year')
df.show()
```
